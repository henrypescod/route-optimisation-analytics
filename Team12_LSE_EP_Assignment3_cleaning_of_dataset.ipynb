### **1 - Order details**

```python
# View the first 5 rows of the dataset.
order_details.head()
```

```python
# Display summary of DataFrame.
order_details.info()
```

```python
# Count the total number of missing values in the DataFrame
null_values_count = order_details.isnull().sum().sum()

# Get columns with missing values
columns_with_null = order_details.columns[order_details.isnull().any()].tolist()

# Format the output message
output_message = f"The total number of null values in the DataFrame is: {null_values_count}\n"
if columns_with_null:
    output_message += f"Missing values found in the following columns: {', '.join(columns_with_null)}"
else:
    output_message += "No columns have null values."

print(output_message)
```

```python
# Look for null values.
order_details.isnull().sum()
```

> There are **171 missing values** in column 'LineWeight'
>
> There are **4124 missing values** in column 'CubicMperUnit'
>
> `Please note that:`**Since the removal of these values could impact the analysis of some of the metrics used, it has been agreed that none of the values will be removed.**

```python
# Rename columns in a more simple manner.
order_details = order_details.rename(columns={
    'OrderID': 'order_id',
    'PartNumber': 'part_number',
    'Description': 'description',
    'Packs': 'packs',
    'PackPrice': 'pack_price',
    'UnitsPerPack': 'units_per_pack',
    'PackWeight': 'pack_weight',
    'LineWeight': 'line_weight',
    'PalletQuantity': 'pallet_quantity',
    'CubicMPerUnit': 'cubic_m_per_unit'
})

# See updated DataFrame.
order_details.info()
```

>### **Add additional columns**

>>**total_pack_price:**

```python
# Calculate 'total_pack_price'.
order_details ['total_pack_price'] = order_details ['packs']* order_details['pack_price']
# See updated DataFrame.
order_details.head()
```

>>**total_units:**

```python
# Calculate 'total_units'
order_details ['total_units'] = order_details ['packs']* order_details['units_per_pack']
# See updated DataFrame.
order_details.head()
```

>>**pallet_space:**

```python
# Calculate 'pallet_space'
# Use apply() to set 'pallet_space' based on a condition
order_details['pallet_space'] = order_details.apply(lambda row: 
                                                    row['total_units'] / row['pallet_quantity']\
                                                    if row['pallet_quantity'] != 0 else 0,axis=1)

# See updated DataFrame.
order_details.head()
```

```python
# Display updated DataFrame.
order_details.info()
```

>>**Recalculate column 'line_weight':**

```python
# Calculate 'Line_weight'
order_details['line_weight_recalc'] = order_details ['packs']* order_details['pack_weight']

# See updated DataFrame.
order_details.head()
```

```python
# Display updated DataFrame.
order_details.info()
```

```python
# Delete unnessary columns
order_details = order_details.drop(['line_weight'], axis= 1)

# View Dataframe
order_details.info()
```

>>**total_cubic_m:**

```python
# Calculate 'total_cubic_m'.
order_details['total_cubic_m'] = order_details ['total_units']* order_details['cubic_m_per_unit']

# See updated DataFrame.
order_details.head()
```

```python
# View Dataframe
order_details.info()
```

```python
# Look for null values.
order_details.isnull().sum()
```

```python
# Save DataFrame as csv.
order_details.to_csv('order_details_clean.csv')
```

### **2 - Orders**

```python
# View the first few rows of the dataset.
orders.head()
```

```python
# Display summary of DataFrame.
orders.info()
```

```python
# Count the total number of missing values in the DataFrame
null_values_count = orders.isnull().sum().sum()

# Get columns with missing values
columns_with_null = orders.columns[orders.isnull().any()].tolist()

# Format the output message
output_message = f"The total number of null values in the DataFrame is: {null_values_count}\n"
if columns_with_null:
    output_message += f"Missing values found in the following columns: {', '.join(columns_with_null)}"
else:
    output_message += "No columns have null values."

print(output_message)
```

```python
# Look for null values.
orders.isnull().sum()
```

> There are **1057 null values** in column 'DateDespatched'
>
> There are **932 null values** in column 'DeliveryCost'

```python
# List of date columns you want to convert
date_columns = ['OrderDate', 'DateDespatched']

# Apply pd.to_datetime to each column in the list
for col in date_columns:
    orders[col] = pd.to_datetime(orders[col], format='%d/%m/%Y %H:%M:%S', errors='coerce')
    
# View Info
orders.info()
```

```python
# Normalize the datetime column to keep only the date part (set time to 00:00:00)
orders['OrderDate'] = pd.to_datetime(orders['OrderDate']).dt.normalize()
```

```python
# Return the status of orders.
orders['StatusName'].unique().tolist()
```

```python
# Return the number of rows with status 'Deleted'.
print("There are", orders[orders.StatusName == 'Deleted'].shape[0],"rows with status 'Deleted'")
```

```python
# Delete orders with status 'Deleted'.
orders = orders[orders['StatusName'] != 'Deleted']

# Reset the index
orders.reset_index(drop=True, inplace=True)
```

```python
# Look for null values.
orders.isnull().sum()
```

> There are now **134 null values** in column 'DateDespatched'.
>
> There are **132 null values** in column 'DeliveryCost'.
> 
>`Please note that:`**These values will be removed from the dataset.**

```python
# View metadata
orders.info()
```

```python
# Delete null values from column 'DateDespatched'
orders = orders.dropna(subset=['DateDespatched'])

# Reset the index
orders.reset_index(drop=True, inplace=True)
```

```python
# Look for null values.
orders.isnull().sum()
```

```python
# View metadata
orders.info()
```

```python
# Rename columns
orders = orders.rename(columns={
    'OrderID': 'order_id',
    'WebsiteID': 'website_id',
    'OrderDate': 'order_date',
    'StatusName': 'status',
    'StatusId': 'status_id',
    'PaymentOptionName': 'payment_option',
    'PaymentType': 'payment_type',
    'DespatchedCourier': 'courier',
    'ProductsExVAT': 'products_ex_vat',
    'DeliveryExVAT': 'delivery_ex_vat',
    'VAT': 'vat',
    'SubTotal': 'subtotal',
    'GrandTotal': 'total',
    'DateDespatched': 'date_despatched',
    'DeliveryCost': 'delivery_cost'
})

# See updated DataFrame.
orders.info()
```

>>**Recalculate column 'subtotal':**

```python
# Calculate 'subtotal'
orders['total_price'] = orders ['products_ex_vat'] + orders['delivery_ex_vat']

# See updated DataFrame.
orders.head()
```

```python
# See updated DataFrame.
orders.info()
```

```python
# Delete unnessary columns
orders = orders.drop(['subtotal'], axis= 1)

# View Dataframe
orders.info()
```

```python
# Save DataFrame as csv.
orders.to_csv('orders_clean.csv')
```

```python
# View the first few rows of the dataset.
customers.head()
```

```python
# Display summary of DataFrame.
customers.info()
```

```python
# Count the total number of missing values in the DataFrame
null_values_count = customers.isnull().sum().sum()

# Get columns with missing values
columns_with_null = customers.columns[customers.isnull().any()].tolist()

# Format the output message
output_message = f"The total number of null values in the DataFrame is: {null_values_count}\n"
if columns_with_null:
    output_message += f"Missing values found in the following columns: {', '.join(columns_with_null)}"
else:
    output_message += "No columns have null values."

print(output_message)
```
